```python
from google.colab import files
uploaded = files.upload()
pdf_name = list(uploaded.keys())[0]
```

```python
from pdf2image import convert_from_path

pages = convert_from_path(pdf_name)

image_files = []

for i, page in enumerate(pages):
    img_name = f"page_{i+1}.png"
    page.save(img_name, "PNG")
    image_files.append(img_name)

print("Converted to images:", image_files)
```

```python
html = """
<!DOCTYPE html>
<html>
<head>
<meta charset="UTF-8">
<title>Professional FlipBook</title>

<style>
body {
    margin: 0;
    background: #222;
    overflow: hidden;
    display: flex;
    justify-content: center;
    align-items: center;
    height: 100vh;
}

.book {
    width: 90vw;
    height: 95vh;
    position: relative;
    perspective: 2000px;
}

.page {
    width: 100%;
    height: 100%;
    position: absolute;
    transform-origin: left;
    transition: transform 0.8s ease-in-out;
    backface-visibility: hidden;
}

.page img {
    width: 100%;
    height: 100%;
    object-fit: contain;
}

.flipped {
    transform: rotateY(-180deg);
}

.controls {
    position: fixed;
    bottom: 20px;
    display: flex;
    gap: 20px;
}

button {
    padding: 10px 20px;
    font-size: 18px;
    cursor: pointer;
}
</style>
</head>

<body tabindex="0">

<div class="book" id="book">

    <!-- Reverse Order -->
    <div class="page"><img src="page_7.png"></div>
    <div class="page"><img src="page_6.png"></div>
    <div class="page"><img src="page_5.png"></div>
    <div class="page"><img src="page_4.png"></div>
    <div class="page"><img src="page_3.png"></div>
    <div class="page"><img src="page_2.png"></div>
    <div class="page"><img src="page_1.png"></div>

</div>

<div class="controls">
    <button onclick="prevPage()">Prev</button>
    <button onclick="nextPage()">Next</button>
    <button onclick="toggleFull()">Full Screen</button>
</div>

<script>

let pages = document.querySelectorAll(".page");
let current = pages.length - 1;

// 🎵 REALISTIC MP3 SOUND
let flipSound = new Audio("flip.mp3");
flipSound.volume = 0.7; // adjust volume here (0.0 to 1.0)

function playFlipSound() {
    flipSound.currentTime = 0; // restart sound if clicked fast
    flipSound.play();
}

function nextPage() {
    if (current >= 0) {
        pages[current].classList.add("flipped");
        current--;
        playFlipSound();
    }
}

function prevPage() {
    if (current < pages.length - 1) {
        current++;
        pages[current].classList.remove("flipped");
        playFlipSound();
    }
}

function toggleFull() {
    if (!document.fullscreenElement) {
        document.documentElement.requestFullscreen();
    } else {
        document.exitFullscreen();
    }
}

// Keyboard Support
document.addEventListener("keydown", function(e) {

    if (e.code === "ArrowRight") {
        e.preventDefault();
        nextPage();
    }

    if (e.code === "ArrowLeft") {
        e.preventDefault();
        prevPage();
    }

});

// Mouse Wheel Support
document.addEventListener("wheel", function(e) {
    if (e.deltaY > 0) nextPage();
    else prevPage();
});

// Focus body for keyboard
window.onload = function() {
    document.body.focus();
};

</script>

</body>
</html>
"""

with open("Professional_FlipBook.html", "w") as f:
    f.write(html)

print("Professional FlipBook created!")
```
